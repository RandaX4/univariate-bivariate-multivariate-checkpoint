```python
from sklearn.preprocessing import StandardScaler
features=["Status","Seniority","Home","Time","Age","Marital","Records","Expenses","Income","Assets","Amount","Price","Finrat","Savings"]
x = df.loc[:,features].values
x_standardized=StandardScaler().fit_transform(x)
x_standardized
```

```python
import numpy as np
cov_df=np.corrcoef(x.T)
cov_df
```

```python
import matplotlib.pyplot as plt
img = plt.matshow(cov_df, cmap=plt.cm.rainbow)
plt.colorbar(img, ticks = [-1, 0, 1], fraction=0.045)
for x in range(cov_df.shape[0]):
    for y in range(cov_df.shape[1]):
        plt.text(x, y, "%0.2f" % cov_df[x,y], size=12, color='black', ha="center", va="center")
        
plt.show()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
principal_components=pca.fit_transform(x_standardized)
principalDF=pd.DataFrame(data=principal_components,columns=['principal component 1','principal component 2'])
finalDf=pd.concat([principalDF,df[['Job']]],axis =1)
finalDf
```

```python
dj=df["Job"]
dj
```

```python
dj.value_counts()
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1',fontsize=15)
ax.set_ylabel('Principal Component 2',fontsize=15)
ax.set_title('2 component PCA',fontsize=20)
targets = ['fixed','freelance','partime','others']
colors = ['r','g','b','y']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Job'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
import prince
mca = prince.MCA(n_components=2,n_iter=3,copy=True,check_input=True,engine='auto',random_state=42)
mca = mca.fit(data_cat)
mca

```
